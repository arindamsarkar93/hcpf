```python
import numpy as np
```

```python
x = np.loadtxt('Y_train.txt',delimiter=',',dtype=np.int8)
```

```python
prob = 0.8
x_train_mask = np.ones(x.shape)
x_test_mask = np.zeros(x.shape)
count = count1 = 0
```

```python
for i in range(0,x.shape[0]):
    for j in range(0,x.shape[1]):
        if x[i,j] == 1.0:
            count1 += 1
            temp = np.random.uniform(size=1)
            if temp >= 0.8:
                x_train_mask[i,j] = 0
                x_test_mask[i,j] = 1 
                count += 1
print count
print count1
```

```python
req = 50
ind = []
index_array = np.arange(x.shape[1])
for i in range(0,x.shape[0]):
    count = 0;
    for j in range(0,x.shape[1]):
        if x_test_mask[i,j] == 1:
            ind.append(j)
            count+=1
    temp = np.array(ind)
    del ind[:]
    temp = np.setdiff1d(index_array,temp,assume_unique=True)
    temp = np.random.choice(temp,size=req-count,replace=False)
    for k in range(0,req-count):
        x_test_mask[i,temp[k]] = 1
        x_train_mask[i,temp[k]] = 0
    print count  
```

```python
for i in range(0,x.shape[0]):
    count = 0;
    for j in range(0,x.shape[1]):
        if x_test_mask[i,j] == 1:
            count += 1
    print count
```

```python
np.savetxt("y_test_mask.txt",x_test_mask,fmt="%d")
np.savetxt("y_train_mask.txt",x_train_mask,fmt="%d")
```
